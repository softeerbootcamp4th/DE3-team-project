```python
# !aws s3 cp s3://de3-web-scraping/ ./ --recursive

import pandas as pd
import glob
import pandas as pd
import matplotlib.pyplot as plt

def read_csv(folder_path):
    # 모든 CSV 파일 경로를 리스트로 가져오기
    all_files = glob.glob(folder_path)
    # 모든 CSV 파일을 읽어서 리스트에 저장
    dfs = []
    for file in all_files:
        # 파일을 텍스트 모드로 읽기
        with open(file, 'r', encoding='utf-8') as file:
            df = pd.read_csv(file, sep=",")  # 각 CSV 파일을 읽음
            dfs.append(df)  # 데이터프레임 리스트에 추가
    df = pd.concat(dfs, ignore_index=True)
    return df


df = pd.concat([read_csv('./clien/코나 화재/posts/*.csv'), read_csv('./fmkorea/코나 화재/posts/*.csv')], ignore_index=True)
df
```

```python
# 다양한 날짜 형식 처리 함수
def parse_dates(date_str):
    return pd.to_datetime(date_str, errors='coerce')

# 'created_at' 열을 datetime 형식으로 변환
df['created_at'] = df['created_at'].apply(parse_dates)

# 시간별로 게시글 수 세기
df.set_index('created_at', inplace=True)
post_counts = df.resample('D').count()['id']  # 일별로 'id'의 개수 세기

# 누적 합계 계산
cumulative_counts = post_counts.cumsum()

# 시각화
plt.figure(figsize=(10, 5))
plt.plot(cumulative_counts.index, cumulative_counts.values, marker='o')
plt.title('누적 게시글 생성 수')
plt.xlabel('날짜')
plt.ylabel('누적 게시글 수')
plt.xticks(rotation=45)
plt.grid()
plt.tight_layout()
plt.show()
```
